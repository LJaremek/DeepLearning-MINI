```python
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
with open("teacher_results.json", "r") as file:
    data = json.load(file)

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["epoch"], df["accuracy"], marker="o")
plt.title("Teacher model - Accuracy in Successive Epochs")
plt.xlabel("Epoch")
plt.ylabel("Accuracy [%]")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["epoch"], df["loss"], marker="o", color="red")
plt.title("Teacher model - Loss in Successive Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df["training_time"]/60)
plt.title("Teacher model - Training Time in seconds per Epoch")
plt.ylabel("Training Time [minutes]")
# plt.grid(True)
plt.xticks([])
plt.show()
```
